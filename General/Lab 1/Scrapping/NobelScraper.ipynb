```python
import requests
from bs4 import BeautifulSoup
import csv 
```

```python
url= "https://www.nobelprize.org/prizes/lists/all-nobel-prizes/"
# Extracting the contents of the webpage to a soup
r = requests.get(url)
c = r.content
soup = BeautifulSoup(c,"html.parser")
print(soup.prettify)
```

```python
data= soup.findAll("article",{"class":"entry"})
output=[]
for i in data: 
    p = i.findAll('p')
    for j in p: 
        work = j.text          
        start_work = work.find('â') + 1     
        work = work[start_work :-1]
        a = j.findAll('a')
        for a_element in a:          
            s = str(a_element)          #converts a_element to string
            w1 = s.find('prizes/')   
            w2 = s.find('/20')
            #print(w2)
            #print(s[w2-2])
            name = a_element.text       #extracts elements that are text in within the a tag 
            field = s[w1+7:w2]          #slice the string to return just the field
            year = s[w2+1:w2+5]
            if s[w2-2]=='e':
                w1= s.find('prizes/')
                w2= s.find('/laureates')
                year= s[w2+11:w2+15]
                field = s[w1+7:w2]
                name = a_element.text
            if w2==-1:
                w1= s.find('prizes/')
                w2= s.find('/laureates')
                year= s[w2+11:w2+15]
                field = s[w1+7:w2]
                name = a_element.text
            print(year,field,name,work)
            output.append([year, field, name, work])
            
        
    
   
```
